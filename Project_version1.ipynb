```python
import pandas as pd
import numpy as np
df = pd.read_excel('intro_final.xlsx')
df.drop(['Name'], axis=1, inplace=True)
df.drop(['Eng_test'], axis=1, inplace=True)
df.head()
```

```python
# new data frame with split value columns 
new = df["Year"].str.split(" ", n = 1, expand = True) 
  
# making separate first name column from new data frame 
df["Grad_Year"]= new[1] 

  
# Dropping old Name columns 
df.drop(columns =["Year"], inplace = True) 
df.head()
```

```python
new = df["work_ex"].str.split(" ", n = 1, expand = True) 
  
# making separate first name column from new data frame 
df["Work_Month"]= new[0] 
# Dropping old Name columns 
df.drop(columns =["work_ex"], inplace = True) 
df.head()
```

```python
df['Undergrad_score'] = df['Undergrad_score'].astype(str)
```

```python
df = df[~df.Undergrad_score.str.contains("CGPA")]
#df = df[~df['your column'].isin(['list of strings'])]
```

```python
df['Undergrad_score'] = df['Undergrad_score'].astype(float)
```

```python
keep = ['Admit', 'Reject']
df = df[df.Status.isin(keep)]
df.head()
```

```python
df.dropna(inplace=True)
df.shape
df.to_csv('Final')
```
